# 輸入資料

input()：不提供提示訊息，要求使用者輸入一行資料。
#input('提示訊息')：提供一個提示訊息，並要求使用者輸入一行資料

```python
# 【練習】設計一個輸入個人資料的程式
# 請輸入姓名：江政欽
# 請輸入年齡：60
# 請輸入身高：173

# 輸出===》使用者資料： 江政欽 身高為 173.0 年齡已經 60
name = input('請輸入姓名：')
age = int(input('請輸入年齡：'))
height = float(input('請輸入身高：'))
print('使用者資料：', name, '身高為', height,'年齡已經', age)
```

《請注意》 input()
#input()會讀入整行資料後形成一個字串傳回，若該行資料中有多個資料，則須進行資料分離。

# 從輸入資料行中分離個別資料

strip()：去除資料行前後的空白。
#split(分隔字元)：將資料行依據分隔字元切割。

```python
# 【練習】請要求使用者將姓名、年齡與身高輸入在一行中，並從中分離出此三項資料。
# 請輸入姓名、年齡、身高，並用空白或逗號隔開：江帥哥, 46 174.5

# 輸出===》使用者資料： 江帥哥 身高為 174.5 年齡已經 46
data = input('請輸入姓名、年齡、身高，並用空白或逗號隔開：')
data = data.strip().replace(',', ' ').split()
name, age, height = data[0], int(data[1]), float(data[2])
print('使用者資料：', name, '身高為', height,'年齡已經', age)
```

# 指定分隔輸出與指定行尾輸出
print(sep='XXX')：指定輸出多筆資料間的分隔號為XXX（預設為空白）。
#print(end='XXX')：指定最後一筆資料輸出後要附加XXX（預設為換行）。

```python
# 【練習】王五同學期中考五科國文、英文、數學、化學、物理的成績資料分別是A, B, C, D, E，請分別用以下格式輸出這些成績資料。
# A+++B+++C+++D+++E
# A+++B---C====D##E
# 例如：

# A, B, C, D, E = [60, 99, 75, 90, 84]

# 60+++99+++75+++90+++84
# 60+++99---75====90++84
A, B, C, D, E = [60, 99, 75, 90, 84]
# 格式一
print(A, B, C, D, E, sep='+++')
# 格式二
print(A, end='+++')
print(B, end='---')
print(C, end='====')
print(D, end='++')
print(E)
```

# %格式化輸出法
%Xd： X表示格位保留，d表示整數輸出，若X前面加0則代表前方空格補上0，但若X前方加-號則代表資料在保留格位內靠左對齊。
#%X.Yf：X表示格位保留（含整數位數、小數點和小數位數），Y表示小數位數保留，f表示小數輸出，若X前面加0則代表前方空格補上0，但若X前方加-號則代#表資料在保留格位內靠左對齊。
#%Xs：X表示格位保留，s表示字串輸出，但若X前方加-號則代表資料在保留格位內靠左對齊。


```python
# 【練習】張三同學成績如下所列：50, 60, 45, 99, 74，請算出其總分以及平均分數，各科成績資料請以四個格位輸出，總分以五個格位輸出，平均分數的小數則以八個格位輸出，並保留至小數第二位。
# 例如

# 張三   --->
#   50  60  45  99  74  328   65.60
name = '張三'
sc = [50, 60, 45, 99, 74]
total = sum(sc)
avg = total/len(sc)
print('%-6s--->'%name)
print('%4d%4d%4d%4d%4d%5d%8.2f'%(sc[0],sc[1],sc[2],sc[3],sc[4],total,avg))
```

```python
# 【練習】張三同學（學號1號）成績如下所列：50, 60, 45, 99, 74，李大同同學（學號2號）成績如下所列：63, 93, 84, 65, 88 請算出其總分以及平均分數，並製作一份成績但報表格式如下：
# 姓名：五個格位
# 學號：五個格位
# 各科成績資料：四個格位
# 總分以五個格位輸出
# 平均分數：六個格位輸出，並保留至小數第二位。
# 例如：

# ------------------------------------------------
# 張三  |00001|  50  60  45  99  74|  328| 65.60
# ------------------------------------------------
# 李大同  |00002|  63  93  84  65  88|  393| 78.60
# ------------------------------------------------
sc = [50, 60, 45, 99, 74]
total = sum(sc)
avg = total/len(sc)
print('-'*48)
print('張三\t|%05d|%4d%4d%4d%4d%4d|%5d|%6.2f'%(1,sc[0],sc[1],sc[2],sc[3],sc[4],total,avg))
print('-'*48)
sc = [63, 93, 84, 65, 88]
total = sum(sc)
avg = total/len(sc)
print('李大同\t|%05d|%4d%4d%4d%4d%4d|%5d|%6.2f'%(2,sc[0],sc[1],sc[2],sc[3],sc[4],total,avg))
print('-'*48)



```

# format()格式化輸出法
格式控制字串序列.format(資料序列)
#格式控制字串序列：'{資料編號:格式代碼}...{資料編號:格式代碼}...'

```python
# 【練習】張三同學成績如下所列：50, 60, 45, 99, 74，請算出其總分以及平均分數，各科成績資料請以四個格位輸出，總分以五個格位輸出，平均分數的小數則以八個格位輸出，並保留至小數第二位。 例如
# 張三   --->
#   50  60  45  99  74  328   65.60
name = '江政欽'
sc = [50, 60, 45, 99, 74]
total = sum(sc)
avg = total/len(sc)
print('{0:6s}--->'.format(name))
print('{0:4d}{1:4d}{2:4d}{3:4d}{4:4d}{5:5d}{6:8.2f}'.format(sc[0],sc[1],sc[2],sc[3],sc[4],total,avg))
```

```python
# 【練習】張三同學（學號1號）成績如下所列：50, 60, 45, 99, 74，李大同同學（學號2號）成績如下所列：63, 93, 84, 65, 88 請算出其總分以及平均分數，並製作一份成績但報表格式如下：
# 姓名：五個格位
# 學號：五個格位
# 各科成績資料：四個格位
# 總分以五個格位輸出
# 平均分數：六個格位輸出，並保留至小數第二位。
# 例如：

# ------------------------------------------------
# 張三  |00001|  50  60  45  99  74|  328| 65.60
# ------------------------------------------------
# 李大同  |00002|  63  93  84  65  88|  393| 78.60
# ------------------------------------------------
name = '張三'
sc = [50, 60, 45, 99, 74]
total = sum(sc)
avg = total/len(sc)
print('-'*48)
print('{8:5s}\t|{0:05d}|{1:4d}{2:4d}{3:4d}{4:4d}{5:4d}|{6:5d}|{7:6.2f}'.format(1,sc[0],sc[1],sc[2],sc[3],sc[4],total,avg,name))
print('-'*48)
name = '李大同'
sc = [63, 93, 84, 65, 88]
total = sum(sc)
avg = total/len(sc)
print('{8:5s}\t|{0:05d}|{1:4d}{2:4d}{3:4d}{4:4d}{5:4d}|{6:5d}|{7:6.2f}'.format(2,sc[0],sc[1],sc[2],sc[3],sc[4],total,avg,name))
print('-'*48)



```
